```python
from instaloader import *
import os
import pathlib
import datetime
```

```python
path = pathlib.Path('data')
profile_names = ["mandalapassion", "mandala_for_all", "beautiful_mandalas"]
L = instaloader.Instaloader()
```

```python
for p_name in profile_names:
    profile = Profile.from_username(L.context, p_name)
    dest = path/p_name
    dest.mkdir(parents=True, exist_ok=True)
    i=0

    for post in profile.get_posts():
        i+=1
        name = str(dest/str(i))
        url = post.url
        L.download_pic(name, url, datetime.datetime.now())
```
